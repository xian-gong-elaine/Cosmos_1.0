```python
df_input_label_indices = pd.read_csv('Datasets/ET23k_Clustering_with_External_Indices.csv')
tech_list = df_input_label_indices['Technology'].tolist()
```

```python
tech_index = 0
info_list = []
```

```python
edge_driver = webdriver.Edge('Datasets/msedgedriver')

while tech_index < len(tech_list):
    print (tech_index)
    row_dict = {}
    tech = tech_list[tech_index]
    print (tech)
    row_dict['Technology'] = tech
    try:
        page = wikipedia.page(tech)
        row_dict['Wikipedia_Content'] = page.content
        row_dict['Wikipedia_Summary'] = page.summary
        row_dict['Wikipedia_External_Links'] = len(page.links)
        try:
            row_dict['Wikipedia_Num_Ref_Links'] = len(page.references)
        except KeyError:
            row_dict['Wikipedia_Num_Ref_Links'] = np.nan        
        try:
            row_dict['Wikipedia_Coordinates'] = page.coordinates
        except KeyError:
            row_dict['Wikipedia_Coordinates'] = np.nan
    except wikipedia.exceptions.PageError:
        try:
            page = wikipedia.page(tech.replace(' ',''))
            row_dict['Wikipedia_Content'] = page.content
            row_dict['Wikipedia_Summary'] = page.summary
            row_dict['Wikipedia_External_Links'] = len(page.links)
            try:
                row_dict['Wikipedia_Num_Ref_Links'] = len(page.references)
            except KeyError:
                row_dict['Wikipedia_Num_Ref_Links'] = np.nan
            try:
                row_dict['Wikipedia_Coordinates'] = page.coordinates
            except KeyError:
                row_dict['Wikipedia_Coordinates'] = np.nan
        except wikipedia.exceptions.PageError:
            row_dict['Wikipedia_Content'] = np.nan
            row_dict['Wikipedia_Summary'] = np.nan
            row_dict['Wikipedia_External_Links'] = np.nan
            row_dict['Wikipedia_Num_Ref_Links'] = np.nan
            row_dict['Wikipedia_Coordinates'] = np.nan
    except wikipedia.exceptions.DisambiguationError:
        row_dict['Wikipedia_Content'] = np.nan
        row_dict['Wikipedia_Summary'] = np.nan
        row_dict['Wikipedia_External_Links'] = np.nan
        row_dict['Wikipedia_Num_Ref_Links'] = np.nan
        row_dict['Wikipedia_Coordinates'] = np.nan        

    url = 'https://en.wikipedia.org/wiki/{0}'.format(tech.replace(' ','_'))
    edge_driver.get(url)
    num_lang = edge_driver.find_element(By.ID,'p-lang-btn-checkbox').get_attribute('aria-label')
    if len(re.findall(r'\d+', num_lang)) != 0:
        num_lang = re.findall(r'\d+', num_lang)[0]
    else:
        num_lang = 0
    print (num_lang)
    row_dict['Wikipedia_Num_Language'] = num_lang
    
    info_list.append(row_dict)
    tech_index = tech_index + 1
    print ('--------------------------------------------------------')
```

```python
tech_index = tech_index + 1
```

```python
df_wiki_meta = pd.DataFrame(info_list)
```

```python
df_wiki_meta['Wikipedia_Num_Language'] = df_wiki_meta['Wikipedia_Num_Language'].apply(lambda x: int(x)+1)
```

```python
df_wiki_meta.to_csv('Datasets/ET23k_Wiki_Metadata.csv', index = False)
```
