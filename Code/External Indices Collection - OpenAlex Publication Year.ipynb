```python
import pandas as pd
import numpy as np
from urllib.request import urlopen
import requests
import json
import time
from urllib.error import URLError
import re
```

```python
# df_valid_tech = pd.read_csv('Datasets/ET30k_Technology_Embeddings_Vectors.csv')
df_valid_tech = pd.read_csv('Datasets/ET23k_Technology_Embeddings_Vectors.csv')
tech_list = df_valid_tech['Technology'].drop_duplicates().tolist()
```

```python
len(tech_list)
```

```python
tech_index = 0
tech_row_list = []
```

```python
while tech_index < len(tech_list):
    print (tech_index)
    tech = tech_list[tech_index]
    tech_edit = tech.replace('–','-')
    tech_edit = re.sub(r'\([^)]*\)', '', tech_edit).strip()
    tech_edit = urllib.parse.quote(tech_edit)
#     tech_edit = tech_edit.replace('%20',' ')
    print (tech)
    print (tech_edit)
    work_api_url = 'https://api.openalex.org/works?search=%22{0}%22&sort=publication_year'.format(tech_edit)
    
    response = urlopen(work_api_url)
    data_json = json.loads(response.read())
    try:
        row_dict = {}
        row_dict['Technology'] = tech
        row_dict['Pub_Counts'] = data_json['meta']['count']
        print (data_json['meta']['count'])
        row_dict['First_Pub_Title'] = data_json['results'][0]['display_name']
        row_dict['First_Pub_Year'] = data_json['results'][0]['publication_year']
        tech_row_list.append(row_dict)

        tech_index = tech_index + 1
    except IndexError:
        tech_index = tech_index + 1
    except UnicodeEncodeError:
        tech_index = tech_index + 1
    print ('==================================================')
```

```python
df_tech_counts_2019_copy = pd.read_csv('Datasets/OpenAlex_Publication_Counts_2019_for_ET23k.csv')
df_tech_counts_2020_copy = pd.read_csv('Datasets/OpenAlex_Publication_Counts_2020_for_ET23k.csv')
df_tech_counts_2021_copy = pd.read_csv('Datasets/OpenAlex_Publication_Counts_2021_for_ET23k.csv')
df_tech_counts_2022_copy = pd.read_csv('Datasets/OpenAlex_Publication_Counts_2022_for_ET23k.csv')
df_tech_counts_2023_copy = pd.read_csv('Datasets/OpenAlex_Publication_Counts_2023_for_ET23k.csv')

df_tech_counts_2019_copy = df_tech_counts_2019_copy[df_tech_counts_2019_copy['Technology'].isin(tech_list)]
df_tech_counts_2020_copy = df_tech_counts_2020_copy[df_tech_counts_2020_copy['Technology'].isin(tech_list)]
df_tech_counts_2021_copy = df_tech_counts_2021_copy[df_tech_counts_2021_copy['Technology'].isin(tech_list)]
df_tech_counts_2022_copy = df_tech_counts_2022_copy[df_tech_counts_2022_copy['Technology'].isin(tech_list)]
df_tech_counts_2023_copy = df_tech_counts_2023_copy[df_tech_counts_2023_copy['Technology'].isin(tech_list)]

df_all_patent_counts = pd.concat([df_tech_counts_2019_copy.rename(columns = {'Pub_Counts':'Pub_Counts_2019'}).set_index('Technology'), 
                                  df_tech_counts_2020_copy.rename(columns = {'Pub_Counts':'Pub_Counts_2020'}).set_index('Technology'),
                                  df_tech_counts_2021_copy.rename(columns = {'Pub_Counts':'Pub_Counts_2021'}).set_index('Technology'),
                                  df_tech_counts_2022_copy.rename(columns = {'Pub_Counts':'Pub_Counts_2022'}).set_index('Technology'),
                                  df_tech_counts_2023_copy.rename(columns = {'Pub_Counts':'Pub_Counts_2023'}).set_index('Technology')], axis = 1)

df_all_patent_counts = df_all_patent_counts[(~df_all_patent_counts['Pub_Counts_2019'].isna()) & (~df_all_patent_counts['Pub_Counts_2020'].isna()) & (~df_all_patent_counts['Pub_Counts_2021'].isna()) & (~df_all_patent_counts['Pub_Counts_2022'].isna()) & (~df_all_patent_counts['Pub_Counts_2023'].isna())]

df_all_patent_counts = df_all_patent_counts[(df_all_patent_counts['Pub_Counts_2019'] != 0) & (df_all_patent_counts['Pub_Counts_2020'] != 0) & (df_all_patent_counts['Pub_Counts_2021'] != 0) & (df_all_patent_counts['Pub_Counts_2022'] != 0) & (df_all_patent_counts['Pub_Counts_2023'] != 0)]

df_all_patent_counts.reset_index().to_csv('Datasets/OpenAlex_Publication_Counts_2019_2023_for_ET23k.csv', index = False)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df_all_patent_counts_copy = df_all_patent_counts.reset_index()[['Technology','Pub_Counts_2019','Pub_Counts_2020','Pub_Counts_2021','Pub_Counts_2022','Pub_Counts_2023']]
df_all_patent_counts_copy.reset_index(inplace = True, drop = True)

three_year_slope_list = []
for i in df_all_patent_counts_copy.index.tolist():
    y_list = df_all_patent_counts_copy[['Pub_Counts_2021','Pub_Counts_2022','Pub_Counts_2023']].iloc[i,:].tolist()
    x_list = [2021,2022,2023]
    if pd.Series(y_list)[pd.Series(y_list).isna()].shape[0] == 0:
        three_year_slope_list.append(np.polyfit(x_list,y_list,1)[0])
    else:
        three_year_slope_list.append(np.nan)
        
df_all_patent_counts_copy['three_year_slope'] = three_year_slope_list
scaler = StandardScaler()
df_all_patent_counts_copy['scaled_three_year_slope'] = scaler.fit_transform(np.array(three_year_slope_list).reshape(-1, 1)).reshape(1,-1).tolist()[0]

five_year_slope_list = []
for i in df_all_patent_counts_copy.index.tolist():
    y_list = df_all_patent_counts_copy[['Pub_Counts_2019', 'Pub_Counts_2020', 'Pub_Counts_2021', 'Pub_Counts_2022', 'Pub_Counts_2023']].iloc[i,:].tolist()
    x_list = [2019,2020,2021,2022,2023]
    if pd.Series(y_list)[pd.Series(y_list).isna()].shape[0] == 0:
        five_year_slope_list.append(np.polyfit(x_list,y_list,1)[0])
    else:
        five_year_slope_list.append(np.nan)

df_all_patent_counts_copy['five_year_slope'] = five_year_slope_list
scaler = StandardScaler()
df_all_patent_counts_copy['scaled_five_year_slope'] = scaler.fit_transform(np.array(five_year_slope_list).reshape(-1, 1)).reshape(1,-1).tolist()[0]

```
