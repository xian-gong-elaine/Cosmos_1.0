```python
edge_driver = webdriver.Edge('Datasets/msedgedriver')

while tech_index < len(tech_list):
    print (tech_index)
    tech = tech_list[tech_index]
    print (tech)
    url = 'https://patents.google.com/?q=(%22{0}%22)&before=filing:20240101&after=filing:20230101'.format(tech.replace(' ','+'))
#     url = 'https://patents.google.com/?q=TI%3d(%22{0}%22)'.format(tech.replace(' ','+'))
    print (url)
    try:
        edge_driver.get(url)
        time.sleep(3)
        patent_num = edge_driver.find_element('xpath', '//*[@id="count"]/div[1]/span[1]/span[3]').text
        if patent_num == '':
            patent_num = 0
        else:
            patent_num = int(patent_num.replace(',','').strip())
    except NoSuchElementException: 
        patent_num = 0    

    print (patent_num)
    print ('-----------------------------------')
    row_dict = {}
    row_dict['Technology'] = tech
    row_dict['Patent_Num_2023'] = patent_num
    row_list.append(row_dict)
    tech_index = tech_index + 1
    print ('================================================================================================')
```

```python
df_patent_count_2019_copy = pd.read_csv('Datasets/ET23k_Google_Patent_Counts_2019.csv')
df_patent_count_2020_copy = pd.read_csv('Datasets/ET23k_Google_Patent_Counts_2020.csv')
df_patent_count_2021_copy = pd.read_csv('Datasets/ET23k_Google_Patent_Counts_2021.csv')
df_patent_count_2022_copy = pd.read_csv('Datasets/ET23k_Google_Patent_Counts_2022.csv')
df_patent_count_2023_copy = pd.read_csv('Datasets/ET23k_Google_Patent_Counts_2023.csv')
```

```python
print (df_patent_count_2019_copy.shape)
print (df_patent_count_2020_copy.shape)
print (df_patent_count_2021_copy.shape)
print (df_patent_count_2022_copy.shape)
print (df_patent_count_2023_copy.shape)
```

```python
df_all_patent_count = pd.concat([df_patent_count_2019_copy.set_index('Technology'),
                                 df_patent_count_2020_copy.set_index('Technology'),
                                 df_patent_count_2021_copy.set_index('Technology'),
                                 df_patent_count_2022_copy.set_index('Technology'),
                                 df_patent_count_2023_copy.set_index('Technology')], axis = 1)

df_all_patent_count = df_all_patent_count[(~df_all_patent_count['Patent_Num_2019'].isna()) & (~df_all_patent_count['Patent_Num_2020'].isna()) & (~df_all_patent_count['Patent_Num_2021'].isna()) & (~df_all_patent_count['Patent_Num_2022'].isna()) & (~df_all_patent_count['Patent_Num_2023'].isna())]


df_all_patent_count = df_all_patent_count[(df_all_patent_count['Patent_Num_2019'] != 0) & (df_all_patent_count['Patent_Num_2020'] != 0) & (df_all_patent_count['Patent_Num_2021'] != 0) & (df_all_patent_count['Patent_Num_2022'] != 0) & (df_all_patent_count['Patent_Num_2023'] != 0)]

df_all_patent_count.shape
```

```python
from sklearn.preprocessing import StandardScaler

df_all_patent_counts_copy = df_all_patent_count.reset_index()[['Technology','Patent_Num_2019','Patent_Num_2020','Patent_Num_2021','Patent_Num_2022','Patent_Num_2023']]
df_all_patent_counts_copy.reset_index(inplace = True, drop = True)

three_year_slope_list = []
for i in df_all_patent_counts_copy.index.tolist():
    y_list = df_all_patent_counts_copy[['Patent_Num_2021','Patent_Num_2022','Patent_Num_2023']].iloc[i,:].tolist()
    x_list = [2021,2022,2023]
    if pd.Series(y_list)[pd.Series(y_list).isna()].shape[0] == 0:
        three_year_slope_list.append(np.polyfit(x_list,y_list,1)[0])
    else:
        three_year_slope_list.append(np.nan)
        
df_all_patent_counts_copy['three_year_slope'] = three_year_slope_list
scaler = StandardScaler()
df_all_patent_counts_copy['scaled_three_year_slope'] = scaler.fit_transform(np.array(three_year_slope_list).reshape(-1, 1)).reshape(1,-1).tolist()[0]

five_year_slope_list = []
for i in df_all_patent_counts_copy.index.tolist():
    y_list = df_all_patent_counts_copy[['Patent_Num_2019', 'Patent_Num_2020', 'Patent_Num_2021', 'Patent_Num_2022', 'Patent_Num_2023']].iloc[i,:].tolist()
    x_list = [2019,2020,2021,2022,2023]
    if pd.Series(y_list)[pd.Series(y_list).isna()].shape[0] == 0:
        five_year_slope_list.append(np.polyfit(x_list,y_list,1)[0])
    else:
        five_year_slope_list.append(np.nan)

df_all_patent_counts_copy['five_year_slope'] = five_year_slope_list
scaler = StandardScaler()
df_all_patent_counts_copy['scaled_five_year_slope'] = scaler.fit_transform(np.array(five_year_slope_list).reshape(-1, 1)).reshape(1,-1).tolist()[0]

```

```python
df_all_patent_counts_copy.sort_values('scaled_five_year_slope', ascending = False)
```
