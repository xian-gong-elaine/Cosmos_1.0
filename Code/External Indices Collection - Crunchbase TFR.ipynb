```python
df_cb = pd.read_csv('Datasets/Global_Companies_by_CB_Rank_(1.4m).csv', low_memory = False)
df_cb_filter = df_cb.copy()
df_cb_filter['Filtering'] = df_cb_filter['Description'].str.lower().tolist()
df_cb_filter = df_cb_filter[~df_cb_filter['Filtering'].isna()]
```

```python
tech_index = 0
tech_tfr_list = []
```

```python
while tech_index < len(tech_list):
    print (tech_index)
    row_dict = {}
    tech = tech_list[tech_index]
    row_dict['Technology'] = tech
    tech_edit = tech.lower()
    tech_edit = tech_edit.replace('–','-')
    tech_edit = re.sub(r'\([^)]*\)', '', tech_edit).strip()
    print (tech_edit)
    row_dict['TFR'] = df_cb_filter[df_cb_filter['Filtering'].str.contains(tech_edit)]['Total Funding Amount Currency (in USD)'].sum()
    tech_tfr_list.append(row_dict)
    tech_index = tech_index + 1
```

```python
df_tech_tfr = pd.DataFrame(tech_tfr_list)
```

```python
df_tech_tfr.sort_values('TFR', ascending = False)
```
