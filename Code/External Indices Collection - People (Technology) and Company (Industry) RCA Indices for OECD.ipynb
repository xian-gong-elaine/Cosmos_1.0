# RCA Measurement

## People RCA

```python
df_global_people_copy = df_global_people.copy()
df_global_people_copy['Global People Counts'] = df_global_people_copy['Global People Counts'].apply(lambda x: int(x.replace(',','')))
df_global_people_copy['World_Ratio'] = df_global_people_copy['Global People Counts']/df_global_people_copy['Global People Counts'].sum()

df_country_people_summ_copy = df_country_people_summ.copy()
df_country_people_summ_copy = df_country_people_summ_copy.replace('No  found',np.nan)
df_country_people_summ_copy = df_country_people_summ_copy.replace('1 result','1')
df_country_people_summ_copy = df_country_people_summ_copy.merge(df_global_people_copy.set_index('Industry'), how = 'left', left_index = True, right_index = True)
for col in df_country_people_summ_copy.columns[:45]:
    df_country_people_summ_copy.loc[~df_country_people_summ_copy[col].isna(),col] = df_country_people_summ_copy.loc[~df_country_people_summ_copy[col].isna(),col].apply(lambda x: int(x.replace(',','')))
    col_ratio = col.split('_')[0] + '_Ratio'
    df_country_people_summ_copy[col_ratio] = df_country_people_summ_copy[col]/df_country_people_summ_copy[col].sum()
    col_rca = col.split('_')[0] + '_RCA'
    df_country_people_summ_copy[col_rca] = df_country_people_summ_copy[col_ratio]/df_country_people_summ_copy['World_Ratio']
    
df_people_rca = df_country_people_summ_copy.loc[:,df_country_people_summ_copy.columns.str.contains('_RCA')]
```

## Company RCA

```python
df_global_company_copy = df_global_company.copy()
df_global_company_copy = df_global_company_copy.replace('No  found',np.nan)
df_global_company_copy = df_global_company_copy.replace('1 result','1')
df_global_company_copy = df_global_company_copy[~df_global_company_copy['Global Company Counts'].isna()]
df_global_company_copy['Global Company Counts'] = df_global_company_copy['Global Company Counts'].apply(lambda x: int(x.replace(',','')))
df_global_company_copy['World_Ratio'] = df_global_company_copy['Global Company Counts']/df_global_company_copy['Global Company Counts'].sum()


df_country_company_summ_copy = df_country_company_summ.copy()
df_country_company_summ_copy = df_country_company_summ_copy.replace('No  found',np.nan)
df_country_company_summ_copy = df_country_company_summ_copy.replace('1 result','1')
df_country_company_summ_copy = df_country_company_summ_copy.loc[df_country_company_summ_copy.index.isin(df_global_company_copy['Industry'])]
df_country_company_summ_copy = df_country_company_summ_copy.merge(df_global_company_copy.set_index('Industry'), how = 'left', left_index = True, right_index = True)
for col in df_country_company_summ_copy.columns[:45]:
    df_country_company_summ_copy.loc[~df_country_company_summ_copy[col].isna(),col] = df_country_company_summ_copy.loc[~df_country_company_summ_copy[col].isna(),col].apply(lambda x: int(x.replace(',','')))
    col_ratio = col.split('_')[0] + '_Ratio'
    df_country_company_summ_copy[col_ratio] = df_country_company_summ_copy[col]/df_country_company_summ_copy[col].sum()
    col_rca = col.split('_')[0] + '_RCA'
    df_country_company_summ_copy[col_rca] = df_country_company_summ_copy[col_ratio]/df_country_company_summ_copy['World_Ratio']
    
df_company_rca = df_country_company_summ_copy.loc[:,df_country_company_summ_copy.columns.str.contains('_RCA')]
```
